```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set visual style
sns.set(style="whitegrid")

# Load the dataset (update with your own path if needed)
df = pd.read_csv("amazon_cleaned.csv")


# --- EDA & Visualization ---

# 1. Distribution of discount percentages
plt.figure(figsize=(10, 5))
sns.histplot(df['discount_percentage'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Discount Percentages")
plt.xlabel("Discount %")
plt.ylabel("Number of Products")
plt.grid(True)
plt.show()

# 2. Average rating distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['rating'], bins=15, kde=True, color='orange')
plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Products")
plt.grid(True)
plt.show()

# 3. Correlation heatmap of numeric features
plt.figure(figsize=(8, 6))
sns.heatmap(df[['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# 4. Top 10 products by rating count
top_rated = df[['product_name', 'rating_count']].sort_values(by='rating_count', ascending=False).dropna().head(10)

top_rated.plot(kind='barh', x='product_name', y='rating_count', figsize=(10, 6), color='purple')
plt.title("Top 10 Most Rated Products")
plt.xlabel("Rating Count")
plt.ylabel("Product Name")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
